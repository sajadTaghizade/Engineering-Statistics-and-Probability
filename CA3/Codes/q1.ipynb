```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(425)
```

1.1

```python
poi_lambda = 10
exp_lambda = 2
geo_p = 0.2
```

```python
var_poi = poi_lambda
var_exp = 1/(exp_lambda * exp_lambda)
var_geo = (1 - geo_p)/(geo_p * geo_p)
mean_poi = poi_lambda
mean_exp = exp_lambda 
mean_geo = 1 / geo_p
print(var_poi,var_exp,var_geo,mean_poi,mean_exp,mean_geo)
```

1.2

```python
size1 = 30
size2 = 300
size3 = 3000
sampling = 1000
poi_means1=[]
poi_means2=[]
poi_means3=[]
exp_means1=[]
exp_means2=[]
exp_means3=[]
geo_means1=[]
geo_means2=[]
geo_means3=[]
for _ in range(sampling):
    poi_means1.append (np.mean(np.random.poisson(lam=poi_lambda, size=size1)))
    poi_means2.append (np.mean(np.random.poisson(lam=poi_lambda, size=size2)))
    poi_means3 .append(np.mean(np.random.poisson(lam=poi_lambda, size=size3)))
    exp_means1 .append (np.mean(np.random.exponential(scale=exp_lambda, size=size1)))
    exp_means2 .append (np.mean(np.random.exponential(scale=exp_lambda, size=size2)))
    exp_means3 .append (np.mean(np.random.exponential(scale=exp_lambda, size=size3)))
    geo_means1 .append (np.mean(np.random.geometric(p=geo_p, size=size1)))
    geo_means2 .append (np.mean(np.random.geometric(p=geo_p, size=size2)))
    geo_means3 .append (np.mean(np.random.geometric(p=geo_p, size=size3)))

plt.figure(figsize=(15, 10))
plt.subplot(3, 3, 1)
plt.hist(poi_means1, bins=50, color='blue')
plt.xlim(8,12)
plt.title('Poisson 30')
plt.subplot(3, 3, 2)
plt.hist(poi_means2, bins=50, color='blue')
plt.xlim(8,12)
plt.title('Poisson 300')
plt.subplot(3, 3, 3)
plt.hist(poi_means3, bins=50, color='blue')
plt.xlim(8,12)
plt.title('Poisson 3000')
plt.subplot(3, 3, 4)
plt.hist(exp_means1, bins=50, color='green')
plt.xlim(1,3)
plt.title('Exponential 30')
plt.subplot(3, 3, 5)
plt.hist(exp_means2, bins=50, color='green')
plt.xlim(1,3)
plt.title('Exponential 300')
plt.subplot(3, 3, 6)
plt.hist(exp_means3, bins=50, color='green')
plt.xlim(1,3)
plt.title('Exponential 3000')
plt.subplot(3, 3, 7)
plt.hist(geo_means1, bins=50, color='red')
plt.xlim(3,7)
plt.title('Geometric 30')
plt.subplot(3, 3, 8)
plt.hist(geo_means2, bins=50, color='red')
plt.xlim(3,7)
plt.title('Geometric 300')
plt.subplot(3, 3, 9)
plt.hist(geo_means3, bins=50, color='red')
plt.xlim(3,7)
plt.title('Geometric 3000')
plt.show()
```

1.3

```python
mean_poisson1 = np.mean(poi_means1)
std_poisson1 = np.std(poi_means1)
error_poisson1 = std_poisson1 / np.sqrt(size1)
mean_poisson2 = np.mean(poi_means2)
std_poisson2 = np.std(poi_means2)
error_poisson2 = std_poisson2 / np.sqrt(size2)
mean_poisson3 = np.mean(poi_means3)
std_poisson3 = np.std(poi_means3)
error_poisson3 = std_poisson3 / np.sqrt(size3)
mean_exp1 = np.mean(exp_means1)
std_exp1 = np.std(exp_means1)
error_exp1 = std_exp1 / np.sqrt(size1)
mean_exp2 = np.mean(exp_means2)
std_exp2 = np.std(exp_means2)
error_exp2 = std_exp2 / np.sqrt(size2)
mean_exp3 = np.mean(exp_means3)
std_exp3 = np.std(exp_means3)
error_exp3 = std_exp3 / np.sqrt(size3)
mean_geo1 = np.mean(geo_means1)
std_geo1 = np.std(geo_means1)
error_geo1 = std_geo1 / np.sqrt(size1)
mean_geo2 = np.mean(geo_means2)
std_geo2 = np.std(geo_means2)
error_geo2 = std_geo2 / np.sqrt(size2)
mean_geo3 = np.mean(geo_means3)
std_geo3 = np.std(geo_means3)
error_geo3 = std_geo3 / np.sqrt(size3)
print(mean_poisson1,std_poisson1,error_poisson1,mean_poisson2,std_poisson2,error_poisson2,mean_poisson3,std_poisson3,error_poisson3)
print(mean_exp1,std_exp1,error_exp1,mean_exp2,std_exp2,error_exp2,mean_exp3,std_exp3,error_exp3)
print(mean_geo1,std_geo1,error_geo1,mean_geo2,std_geo2,error_geo2,mean_geo3,std_geo3,error_geo3)
```

As the sample size increases, the standard error generally decreases. This is because the estimated mean becomes more accurate as the number of data increases. The mean remains approximately constant, but the standard error decreases. This indicates that the variability decreases and the accuracy of the estimate of the mean increases with increasing sample size.

```python
plt.figure(figsize=(15, 10))
plt.subplot(3, 3, 1)
plt.hist(poi_means1, bins=50, color='blue', density=True)
x = np.linspace(min(poi_means1), max(poi_means1), 100)
plt.plot(x, norm.pdf(x, mean_poisson1, std_poisson1), color='black')
plt.title('Poisson 30')

plt.subplot(3, 3, 2)
plt.hist(poi_means2, bins=50, color='blue', density=True)
x = np.linspace(min(poi_means2), max(poi_means2), 100)
plt.plot(x, norm.pdf(x, mean_poisson2, std_poisson2), color='black')
plt.title('Poisson 300')

plt.subplot(3, 3, 3)
plt.hist(poi_means3, bins=50, color='blue', density=True)
x = np.linspace(min(poi_means3), max(poi_means3), 100)
plt.plot(x, norm.pdf(x, mean_poisson3, std_poisson3), color='black')
plt.title('Poisson 3000')

plt.subplot(3, 3, 4)
plt.hist(exp_means1, bins=50, color='green',density=True)
x = np.linspace(min(exp_means1), max(exp_means1), 100)
plt.plot(x, norm.pdf(x, mean_exp1, std_exp1), color='black')
plt.title('Exponential 30')

plt.subplot(3, 3, 5)
plt.hist(exp_means2, bins=50, color='green',density=True)
x = np.linspace(min(exp_means2), max(exp_means2), 100)
plt.plot(x, norm.pdf(x, mean_exp2, std_exp2), color='black')
plt.title('Exponential 300')

plt.subplot(3, 3, 6)
plt.hist(exp_means3, bins=50, color='green',density=True)
x = np.linspace(min(exp_means3), max(exp_means3), 100)
plt.plot(x, norm.pdf(x, mean_exp3, std_exp3), color='black')
plt.title('Exponential 3000')

plt.subplot(3, 3, 7)
plt.hist(geo_means1, bins=50, color='red',density=True)
x = np.linspace(min(geo_means1), max(geo_means1), 100)
plt.plot(x, norm.pdf(x, mean_geo1, std_geo1), color='black')
plt.title('Geometric 30')

plt.subplot(3, 3, 8)
plt.hist(geo_means2, bins=50, color='red',density=True)
x = np.linspace(min(geo_means2), max(geo_means2), 100)
plt.plot(x, norm.pdf(x, mean_geo2, std_geo2), color='black')
plt.title('Geometric 300')

plt.subplot(3, 3, 9)
plt.hist(geo_means3, bins=50, color='red',density=True)
x = np.linspace(min(geo_means3), max(geo_means3), 100)
plt.plot(x, norm.pdf(x, mean_geo3, std_geo3), color='black')
plt.title('Geometric 3000')

plt.tight_layout()
plt.show()
```

It shows that the normal and 3 other distributions almost overlap, so it is possible to estimate the other 3 distributions with a good approximation using the normal distribution.

It has definitely become more similar to a normal distribution with increasing samples, according to the specified shapes.

As it has been shown, it tends to a normal distribution and the parameters of the normal distribution depend on the mean and standard deviation of the initial distributions.
