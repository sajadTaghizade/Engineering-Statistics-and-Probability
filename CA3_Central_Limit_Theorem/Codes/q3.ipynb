```python
n = 270      
p = 0.3      
x = np.arange(0, n)
pmfs = binom.pmf(x, n, p)
plt.bar(x, pmfs, color='blue')
plt.show()
```

```python
n = 270
p = 0.3

mean = n * p
std = np.sqrt(n * p * (1 - p))
x_binom = np.arange(0, n)
pmfs = binom.pmf(x_binom, n, p)
z_binom = (x_binom - mean) / std 
x_norm = np.linspace(-5, 5, 270)
pdf_y = norm.pdf(x_norm, loc=0, scale=1)
plt.bar(z_binom, pmfs , color='blue')
plt.plot(x_norm, pdf_y, color='red')
plt.show()
```

It can be seen that the two normal and 2-term graphs do not overlap. Of course, the reason for this is that in the normal graph, the surface area under the graph is equal to one, but in the 2-term graph, the sum of the columns is equal to one. For this reason, the two graphs do not overlap.
